```python
def menu(step, name=''):

  if step == 'START':
    print('Работа с телефонной кигой. Выбирете пункт меню')
    print('START - вывести меню')
    print('OPEN <имя файла> - считать данные из файла')
    print('FIND <имя> - поиск по имени/телефону/id')
    print('DEL <id> - удаление данных')
    print('ADD - добавление данных (через запятую, без пробелов')
    print('CHANGE <шаблон для поиска> - изменить данные')
    print('END - закончить работу')

  if step == 'OPEN':
    open_file(name)

  if step == 'FIND':
    find_man(name.lower())

  if step == 'DEL':
    del_line(name)

  if step == 'ADD':
    name = input('\nВведите данные в формате:\nid,lastname,firstname,secondname,phone\nЧЕРЕЗ ЗАПЯТЫЕ И БЕЗ ПРОБЕЛОВ\n')
    add_line(name)

  if step == 'CHANGE':
    change_line(name)

  if step == 'END':
    print('До свидания!')
```

```python
def read_phonebook(line):
  if line == 'id,lastname,firstname,secondname,phone\n':
    return

  global phonebook

  line = line.strip().split(',')
  fio = ' '.join(line[1:4])

  if not line[0].isdigit():
    print(line)
    print('Не верный тип id, строка не будет считана\n')
    return

  if not line[1].isalpha() or not line[2].isalpha() or not line[3].isalpha():
    print(line)
    print('В ФИО обнаружены числа, строка не будет считана\n')
    return

  if not line[4].isdigit():
    print(line)
    print('Буквы в телефонном номере, строка не будет считана\n')
    return

  if line[0] not in phonebook:
    phonebook[line[0]] = [fio, line[4]]
```

```python
def open_file(file_name):
  global phonebook
  phonebook['file_name'] = file_name

  try:
    with open(file_name) as f:
      print('Файл найден')
      for line in f:
        read_phonebook(line)

  except FileNotFoundError:
    print('Файл создан')
    with open(file_name, 'w') as f:
      f.write('id,lastname,firstname,secondname,phone\n')

  print('Файл в оперативной памяти')
```

```python
def find_man(man):
  find_dict = dict()
  for el in phonebook:

    if man == el:
      find_dict[el] = phonebook[el]

    else:
      if man in phonebook[el][0].lower() or man == phonebook[el][1]:
        find_dict[el] = phonebook[el]

  if find_dict:
    print('Найдены следующие элементы:')
    print(find_dict)

    return find_dict

  else:
    print('Ничего не найдено')
    return None
```

```python
def del_line(id_line):
  global phonebook

  del(phonebook[id_line])
  print('Данные удалены из памяти')

  with open(phonebook['file_name'], 'r') as f:
    data = f.readlines()

  with open(phonebook['file_name'], 'w') as f:
    for line in data:
      if line.strip().split(',')[0] == id_line:
        continue
      f.write(line)

  print('Данные удалены из файла')
```

```python
def add_line(line):
  global phonebook

  read_phonebook(line)
  print('Данные записаны в память')

  with open(phonebook['file_name'], 'a') as f:
      f.write(line + '\n')

  print('Данные записаны в файл')
```

```python
def change_line(find_str):
  find_lines = find_man(find_str.lower())

  if find_lines:
    line_id = input('Какую строку изменить: ')
    new_line = input('\nВведите новые данные в формате:\nid,lastname,firstname,secondname,phone\nЧЕРЕЗ ЗАПЯТЫЕ И БЕЗ ПРОБЕЛОВ\n')

    del_line(line_id)  
    add_line(new_line)

  else:
    return
```

```python
step = 1
menu('START')
phonebook = dict()

while step != 'END':

  user = input('\nЧто вы хотите сделать: ').split()
  step = user[0].upper()
  
  if len(user) == 2:
    name = user[1]

  else:
    name = ''

  menu(step, name)
```
